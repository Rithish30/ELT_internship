```python
import json
import os
```

```python
def loadsjson():
    #load records.json
    if os.path.exists('records.json'):
        if os.stat("records.json").st_size == 0:
            return {}
        else:
            f=open("records.json","r")
            records=json.loads(f.read())
            return records    
    else:
        return {}
```

```python
def dumpsjson(records):
#to save all the changes made in edited file
        txt=json.dumps(records)
        with open('records.json','w') as fd:
          fd.write(txt)
        print('***************  Record Saved  ***************')
```

```python
def addRecord(idt,records):
    #This method is used for add record if not already present
    c=0
    add=input("Do you want to add new record y/n  ").lower()
    if add=='y':
        records[idt]={}
        name=input("Enter product name  ")
        brand=input("Enter parent company/brand name ")
        age=input("Enter age category: Adult or Kids")
        gender=input("Enter gender")
        category=input("Enter category")
        var=input("Enter product Variety ")
        material=input("Enter product material")
        color=input("Enter product color")
        size=input("Enter product size")
        price=int(input("Enter Product Price  "))
        quant=int(input("Enter Product Quantity  "))
        dis=int(input("Enter Product Discount % "))
        records[idt]['name']=name
        records[idt]['brand']=brand
        records[idt]['age']=age
        records[idt]['gender']=gender
        records[idt]['category']=category
        records[idt]['variety']=var
        records[idt]['color']=color
        records[idt]['size']=size
        records[idt]['price']=price
        records[idt]['quantity']=quant
        records[idt]['discount']=dis
        print("Record added successfully ")
        c+=1
    else:
        print("Operation Cancelled ")
    return c
```

```python
def updateRecord(idt,records):
  #This function/method is used for update Record.
    c=0
    flag=True
    update=input("Do you want to update it? y/n or enter q to exit ").lower()
    if update=='y':
        while True:
            k=input("Which one do you want to update Name/Brand/age/Gender/category/Variety/Material/color/size/Price/Quantity/Discount  ").lower()
            if k in records[idt]:
                records[idt][k]=input(f"Enter new Value for {k}  ")
                c+=1
                break
            else:
                print("Wrong value Please enter Again  ")
        print("***************Record updated successfully***************")
    elif update=='q':
        flag=False
    else:
        print("Operation Cancelled ")
    return c
```

```python
def operations(t,records):
#this function is used to confirm the operation the user wants to perform ie. either add or update
    c=0
    if not t:
        print("you can only add because the records are empty")
        idt=int(input("Enter Product id  "))
        c=addRecord(idt,records)
        
    else:
        while True:
            op=input("Enter 1 to add a record, 2 to update a product, q to exit  ")
            if op=='1':
                idt=int(input("Enter Product id  "))
                if (idt or int(idt)) not in records:
                    c=addRecord(idt,records)
                    
                elif (idt or int(idt)) in records:
                    inp=input("product is already present. Do you want to update it?y/n").lower()
                    if inp=='y':
                        c=updateRecord(idt,records)
                    else:
                        continue
            elif op=='2':
                idt=input("Enter Product id  ")
                c=updateRecord(idt,records)
            elif op=='q':
                return c
            else:
                print("Wrong input, Enter correct value")
                    
```
